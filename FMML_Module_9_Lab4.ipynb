## Using learnt representations
In practice, very few people train an entire Convolutional Network from scratch (with random initialization), because it is relatively rare to have a dataset of sufficient size. Instead, it is common to pretrain a ConvNet on a very large dataset (e.g. ImageNet, which contains 1.2 million images with 1000 categories), and then use the ConvNet either as an initialization or a fixed feature extractor for the task of interest.


We'll train a model to classify ants and bees. We have about 120 training images each for ants and bees. There are 75 validation images for each class. Usually, this is a very small dataset to generalize upon, if trained from scratch. Since we are using transfer learning, we should be able to generalize reasonably well.

This dataset is a very small subset of imagenet.




```python
# Import packages
import torch
import torchvision
import torchvision.transforms as transforms
from torchvision.datasets import ImageFolder
from torch.utils.data import DataLoader, SubsetRandomSampler
import torch.nn as nn
from torch import optim
import torch.nn.functional as F

import matplotlib.pyplot as plt
import numpy as np
import time
import os
```

```python
# Device configuration (whether to run on GPU or CPU)
device = 'cpu' #torch.device('cuda' if torch.cuda.is_available() else 'cpu')
```

```python
# Download and extract dataset
!wget https://download.pytorch.org/tutorial/hymenoptera_data.zip
!unzip -q hymenoptera_data.zip
```

```python
# Data augmentation and normalization for training
# Just normalization for validation
train_transform = transforms.Compose([
        transforms.RandomResizedCrop(224),
        transforms.RandomHorizontalFlip(),
        transforms.ToTensor(),
        transforms.Normalize([0.485, 0.456, 0.406], [0.229, 0.224, 0.225])
    ])
val_transform = transforms.Compose([
        transforms.Resize(256),
        transforms.CenterCrop(224),
        transforms.ToTensor(),
        transforms.Normalize([0.485, 0.456, 0.406], [0.229, 0.224, 0.225])
    ])

data_dir = './hymenoptera_data'
train_dataset = ImageFolder(os.path.join(data_dir, 'train'), train_transform)
val_dataset = ImageFolder(os.path.join(data_dir, 'val'), val_transform)
train_dataloader = DataLoader(train_dataset, batch_size=4,shuffle=True, num_workers=2)
val_dataloader = DataLoader(val_dataset, batch_size=4,shuffle=True, num_workers=2)
class_names = train_dataset.classes
```

```python
def imshow(inp, title=None):
    """Imshow for Tensor."""
    inp = inp.numpy().transpose((1, 2, 0))
    mean = np.array([0.485, 0.456, 0.406])
    std = np.array([0.229, 0.224, 0.225])
    inp = std * inp + mean
    inp = np.clip(inp, 0, 1)
    plt.imshow(inp)
    if title is not None:
        plt.title(title)
    plt.pause(0.001)  # pause a bit so that plots are updated


# Get a batch of training data
inputs, classes = next(iter(train_dataloader))

# Make a grid from batch
out = torchvision.utils.make_grid(inputs)

imshow(out, title=[class_names[x] for x in classes])
```

```python
def train_model(model, train_dataloader, criterion, optimizer, num_epochs=25):

    for epoch in range(num_epochs):
        print(f'Epoch {epoch}/{num_epochs - 1}')
        print('-' * 10)
        model.train()
        running_loss = 0.0
        running_corrects = 0

        # Iterate over data.
        for i, (inputs, labels) in enumerate(train_dataloader):
            inputs = inputs.to(device)
            labels = labels.to(device)

            # zero the parameter gradients
            optimizer.zero_grad()

            # forward
            # track history if only in train
            with torch.set_grad_enabled(True):
              outputs = model(inputs)
              _, preds = torch.max(outputs, 1)
              loss = criterion(outputs, labels)

              # backward + optimize only if in training phase
              loss.backward()
              optimizer.step()

            # statistics
            running_loss += loss.item() * inputs.size(0)
            running_corrects += torch.sum(preds == labels.data)

        epoch_loss = running_loss / len(train_dataloader.dataset)
        epoch_acc = running_corrects.double() / len(train_dataloader.dataset)

        print(f'Loss: {epoch_loss:.4f} Acc: {epoch_acc:.4f}')

        print()
```

```python
def visualize_model(model, num_images=10):
    was_training = model.training
    model.eval()
    images_so_far = 0
    fig = plt.figure()

    with torch.no_grad():
        for i, (inputs, labels) in enumerate(val_dataloader):
            inputs = inputs.to(device)
            labels = labels.to(device)

            outputs = model(inputs)
            _, preds = torch.max(outputs, 1)

            for j in range(inputs.size()[0]):
                images_so_far += 1
                ax = plt.subplot(num_images//2, 2, images_so_far)
                ax.axis('off')
                ax.set_title(f'predicted: {class_names[preds[j]]}')
                imshow(inputs.cpu().data[j])

                if images_so_far == num_images:
                    model.train(mode=was_training)
                    return
        model.train(mode=was_training)
```

```python
from torchvision import models

model_ft = models.resnet18(pretrained=True)
num_ftrs = model_ft.fc.in_features
# Here the size of each output sample is set to 2.
# Alternatively, it can be generalized to nn.Linear(num_ftrs, len(class_names)).
model_ft.fc = nn.Linear(num_ftrs, 2)

model_ft = model_ft.to(device)

loss_func = nn.CrossEntropyLoss()

# Observe that all parameters are being optimized
optimizer = optim.SGD(model_ft.parameters(), lr=0.001, momentum=0.9)
```

```python
num_epochs = 5
train_model(model_ft, train_dataloader, loss_func, optimizer, num_epochs)
```

```python
visualize_model(model_ft)
```

#**ANSWERS**

#**1) QUESTION ANSWER**



Data augmentations like resize, crop, rotation, flipping, etc., serve several important purposes in training data:

* **Increased Robustness**: By applying various transformations to the input data, the model becomes more robust to changes and variations in the input during inference. This helps improve generalization to unseen data.

* **Improved Generalization**: Data augmentations introduce diversity into the training dataset, which helps prevent overfitting by exposing the model to a wider range of variations of the input data.

* **Enhanced Performance**: Augmentations can help improve the performance of the model by providing it with more training examples, even if they are generated from a smaller original dataset.

* **Regularization**: Data augmentations act as a form of regularization by adding noise to the training process, which can help prevent the model from memorizing the training data and encourage it to learn more robust and generalizable features.

Overall, data augmentations are a crucial component of training deep learning models, especially when working with limited datasets, as they help improve the model's ability to learn and generalize from the available data.

#**2) QUESTION ANSWER**


If you don't use a pretrained ResNet model and train it from scratch by setting `pretrained=False`, the performance of the model might be lower compared to using a pretrained model. This is because training a deep neural network from scratch on a limited dataset may require a large amount of data and computational resources to achieve comparable performance.

Without pretrained weights, the model starts with random initialization of its parameters, which means it needs to learn meaningful representations from scratch solely based on the training data. This can be challenging, especially for deep architectures like ResNet, and may result in longer training times and potentially poorer generalization performance.

However, the exact performance you'll get depends on various factors such as the size and diversity of your training dataset, the complexity of the task, the model architecture, the optimization algorithm used, and the training hyperparameters. It's recommended to experiment with different configurations and hyperparameters to achieve the best performance when training a model from scratch.

#**3) QUESTION ANSWER**


Using a ResNet model pretrained on a significantly different dataset may still yield beneficial performance improvements, even when applied to a new task like classifying ants vs. bees. This is because pretraining on a diverse dataset helps the model learn general features and patterns that are transferable across different tasks and domains. These learned features often capture low-level visual representations such as edges, textures, and shapes, which are useful for many computer vision tasks.

However, the extent to which the pretrained model improves performance depends on several factors:

* **Domain Similarity**: The degree of similarity between the pretrained dataset and the target dataset (ants vs. bees) matters. If the pretrained dataset is somewhat related to the target task in terms of visual characteristics, the pretrained model is likely to provide more relevant features and thus better performance.

* **Task Complexity**: The complexity of the target task also plays a role. If the task requires recognizing high-level semantic concepts that are vastly different from those in the pretrained dataset, the benefits of transfer learning may be limited.

* **Fine-Tuning**: Fine-tuning the pretrained model on the target dataset allows the model to adapt its learned features to the specific characteristics of the new dataset. This process can further improve performance, even when the pretrained dataset is different.

* **Amount of Data**: The amount of data available for fine-tuning also influences performance. If the target dataset is small, transfer learning from a pretrained model becomes even more valuable, as it helps mitigate overfitting and leverages the knowledge captured by the pretrained model.

In summary, while using a pretrained model on a significantly different dataset may not guarantee optimal performance, it often provides a good starting point and can significantly accelerate the training process, especially when fine-tuned on the target dataset.
